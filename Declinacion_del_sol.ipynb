# **Declinación del sol en base a la esfera celeste**

De acuerdo a la configuración de la esfera celeste, la declinación de las estrellas en el plano del cielo se mantienen constantes. Sin embargo, la posición de nuestro sol cambia en el sistema ecuatorial a lo largo del año. Esto se debe a la inclinación del eje de rotación de la tierra con respecto al eje perpendicular de la orbita del sol conocido como ecliptica, que es igual a $\varepsilon_0=23.45^\circ$. 

Esta variación esta dada de forma aproximada por la siguiente expresión:

$$  
\delta_{\odot} = -\arcsin \left[ \sin \epsilon_0 \cos \left( \frac{360^\circ}{365.24} (N + 10) \right) \right]
   $$

- Donde $N$ es la diferencia en dias, comenzando desde el 1ro de enero. El primer dia es N=0 y el último N=364 (depende si el año es bisiesto).

- La razón $\frac{360^\circ}{365.24}$ es el cambio de la posicion angular de la Tierra sobre día, asumiendo que la tierra tiene orbita circular. Este es justamente la velocidad angular $\omega$ del movimiento de la Tierra alrededor del Sol, en unidades de grados sobre dia.


El Sol tiene ``declinación cero`` en los equinoccios (puntos de intersección del ecuador celeste y la eclíptica) y alcanza $\pm \varepsilon_0$ en los solsticios, donde el eje de rotación de la Tierra se inclina hacia o lejos del Sol. Las fechas exactas varían un poco de un año a otro.

En 2020, por ejemplo, los equinoccios fueron el 20 de marzo y el 22 de septiembre, y los solsticios el 20 de junio y el 21 de diciembre.

### **Ejemplo de aplicación**

En el siguiente ejemplo, se desea determinar el valor de $\delta_{\odot}$ con el valor correspondiente de N. Por ejemplo, el 20 de junio es el día 172 del año 2020. Contando desde cero, esperamos que el máximo de la declinación $\delta_{\odot} = \varepsilon_0$ (primer solsticio) se produzca para N = 171. Veamos si esto es consistente con la aproximación.

```python
import math

#Numero de día
N=171

#Velocidad angular de la Tierra
w=2*math.pi/365.25

#Valor constante de la epsilon_0
e_0=math.radians(23.44)


#Calculo de la declinación delta

delta=-math.asin(math.sin(e_0)*math.cos(w*(N+10)))
print("La declinación solar es: ", math.degrees(delta), "grados")
```

El valor calculado coincide con $\varepsilon_0$, debido a que se tomo una fecha de solticio, pero si se toma alguna otra, la declinación será diferente.

#### **Indtroduccion del valor de N como lista**

Se puede calcular de forma simultanea distintos valores de $N$ gracias a la estructura de datos del tipo ``lista``. Para esto, se debe utilizar la libreria de ``numpy`` que ayuda a trabajar con arrays o listas, ya que la libreria math solo funciona para calculos numericos exactos. A continuación de presenta la sintaxis:

```python
import math
import numpy as np

#Numeros de día
N=np.array([171, 180, 190])

#Velocidad angular de la Tierra
w=2*math.pi/365.25

#Valor constante de la epsilon_0
e_0=math.radians(23.44)


#Calculo de la declinación delta

delta=-np.arcsin(np.sin(e_0)*np.cos(w*(N+10)))
print("La declinación solar es: ", np.degrees(delta), "grados")
```
