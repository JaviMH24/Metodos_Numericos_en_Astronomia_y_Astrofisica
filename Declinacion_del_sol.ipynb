# **Declinación del sol en base a la esfera celeste**

De acuerdo a la configuración de la esfera celeste, la declinación de las estrellas en el plano del cielo se mantienen constantes. Sin embargo, la posición de nuestro sol cambia en el sistema ecuatorial a lo largo del año. Esto se debe a la inclinación del eje de rotación de la tierra con respecto al eje perpendicular de la orbita del sol conocido como ecliptica, que es igual a $\varepsilon_0=23.45^\circ$. 

Esta variación esta dada de forma aproximada por la siguiente expresión:

$$  
\delta_{\odot} = -\arcsin \left[ \sin \epsilon_0 \cos \left( \frac{360^\circ}{365.24} (N + 10) \right) \right]
   $$

- Donde $N$ es la diferencia en dias, comenzando desde el 1ro de enero. El primer dia es N=0 y el último N=364 (depende si el año es bisiesto).

- La razón $\frac{360^\circ}{365.24}$ es el cambio de la posicion angular de la Tierra sobre día, asumiendo que la tierra tiene orbita circular. Este es justamente la velocidad angular $\omega$ del movimiento de la Tierra alrededor del Sol, en unidades de grados sobre dia.


El Sol tiene ``declinación cero`` en los equinoccios (puntos de intersección del ecuador celeste y la eclíptica) y alcanza $\pm \varepsilon_0$ en los solsticios, donde el eje de rotación de la Tierra se inclina hacia o lejos del Sol. Las fechas exactas varían un poco de un año a otro.

En 2020, por ejemplo, los equinoccios fueron el 20 de marzo y el 22 de septiembre, y los solsticios el 20 de junio y el 21 de diciembre.

### **Ejemplo de aplicación**

En el siguiente ejemplo, se desea determinar el valor de $\delta_{\odot}$ con el valor correspondiente de N. Por ejemplo, el 20 de junio es el día 172 del año 2020. Contando desde cero, esperamos que el máximo de la declinación $\delta_{\odot} = \varepsilon_0$ (primer solsticio) se produzca para N = 171. Veamos si esto es consistente con la aproximación.

```python
import math

#Numero de día
N=171

#Velocidad angular de la Tierra
w=2*math.pi/365.25

#Valor constante de la epsilon_0
e_0=math.radians(23.44)


#Calculo de la declinación delta

delta=-math.asin(math.sin(e_0)*math.cos(w*(N+10)))
print("La declinación solar es: ", math.degrees(delta), "grados")
```

El valor calculado coincide con $\varepsilon_0$, debido a que se tomo una fecha de solticio, pero si se toma alguna otra, la declinación será diferente.

#### **Indtroduccion del valor de N como lista**

Se puede calcular de forma simultanea distintos valores de $N$ gracias a la estructura de datos del tipo ``lista``. Para esto, se debe utilizar la libreria de ``numpy`` que ayuda a trabajar con arrays o listas, ya que la libreria math solo funciona para calculos numericos exactos. A continuación de presenta la sintaxis:

```python
import math
import numpy as np

#Numeros de día
N=np.array([171, 180, 190])

#Velocidad angular de la Tierra
w=2*math.pi/365.25

#Valor constante de la epsilon_0
e_0=math.radians(23.44)


#Calculo de la declinación delta

delta=-np.arcsin(np.sin(e_0)*np.cos(w*(N+10)))
print("La declinación solar es: ", np.degrees(delta), "grados")
```

### **Mostrar los resultados en forma de tabla**

Para mostar los resultados en formato de ciclo, debemos generar un for del siguiente modo. La variable val es ua variable auxiliar que nos ayudará a movernos por la lista delta

```python
for val in delta:
    print("La declinación solar es: ", np.degrees(val), "grados")
```

Si bien imprimir repetidamente la declinación es redundante, falta información importante para comprender los datos. En particular, no se especifican los días a los que se refieren estos valores. Una mejor manera de imprimir datos relacionados en matrices es, por supuesto, una ``tabla``.


Para imprimir los días y las declinaciones correspondientes, necesitamos iterar simultáneamente a través de los elementos de ``N`` y ``delta``. Una solución es definir un contador con la ayuda de la función ``enumerate()`` de Python:


```python
print("i  day  delta [deg]")
for i,val in enumerate(delta):
    print("{1:d} {2:3d}  {0:8.2f}".format(np.degrees(val), i, N[i]))
```

El marcador de posición ``{1:d}`` indica que el segundo argumento (índice 1 antes de los dos puntos) se insertará como un entero en esta posición. El siguiente marcador de posición se refiere al tercer argumento (índice 2 antes de los dos puntos) y el último al primero.

Los especificadores de formato aseguran suficiente espacio entre las columnas de números.

### **Crear una tabla sin un contador**

Una laterniatva es que, en lugar de enumerar ``delta``, podría enumerar ``N``. ¿Es posible escribir un bucle que itere simultáneamente ambos arreglos sin usar un contador explícito?

Para esto, se utiliza la función ``zip()``, que agrega elementos con el mismo índice de dos o más arreglos de igual longitud.

En este caso, la variable del bucle es una tupla que contiene un elemento de cada arreglo. La tupla `row` en el siguiente ejemplo de código contiene un elemento `N` y el elemento `delta` correspondiente, formando así una fila de la tabla que queremos imprimir. Los elementos de la tupla se indexan como los de un array, por lo que `row[0]` se refiere al día y `row[1]` a la declinación de ese día.

```python
print("day delta [deg]")
for row in zip(N, delta):
    print("{0:3d}  {1:8.2f}".format(row[0], np.degrees(row[1])))
```

Si reemplazamos el ``print("{0:3d}  {1:8.2f}".format(row[0], np.degrees(row[1])))`` por ``print(row)`` se imprimiran en forma de tuplas los valores calulados:

```python
print("day delta [deg]")
for row in zip(N, delta):
    print(row)
```

### **Consideraciones pertinentes**

Mientras que los corchetes se usan para listas y matrices, las tuplas se encierran entre paréntesis. La diferencia más importante es que las tuplas son ``inmutables``; es decir, no es posible agregar, cambiar ni eliminar elementos.

Las tuplas mostradas arriba son pares fijos de números, similares a las coordenadas ``(x, y)`` de puntos en un plano.

Supongamos que queremos cambiar el día 355 por el 364. Si bien es posible modificar la matriz N estableciendo ``N[3] = 364``, no se permite asignar un valor a un elemento individual de una tupla, como ``row[0]``. Solo se puede sobrescribir la tupla completa con una nueva (esto es lo que sucede en el bucle anterior).
