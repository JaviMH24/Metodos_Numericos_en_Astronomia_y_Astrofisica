# **Observación de objetos celestes**

Aunque podemos especificar las posiciones de los objetos celestes en un sistema de coordenadas independiente del tiempo y de la ubicación del observador, en definitiva, lo que se desea es saber dónde se encuentra el objeto en el cielo.

Para ello, se utiliza el llamado ``sistema de coordenadas horizontales``, que se basa en un plano imaginario orientado tangente a la superficie de la Tierra en la ubicación del observador, por lo que dos personas en distintos lugares verán coordenadas diferentes para el mismo astro al mismo tiempo.

#### Puntos clave:

**Cenit (Z)**: Punto directamente sobre la cabeza del observador.

**Nadir (N)**: Punto directamente bajo sus pies.

**Horizonte**: Círculo máximo que divide la esfera celeste en hemisferio visible y no visible.

#### Coordenadas principales

**Altura (h o a)**: Este es un ángulo entre el astro y el horizonte. Se mide de 0° (en el horizonte) a 90° (en el cenit).

**Acimut (Az)**: Ángulo sobre el horizonte contado desde un punto cardinal (generalmente el Norte en astronomía, aunque en navegación puede ser el Sur). Se mide en sentido horario de 0° a 360°. Para cada observador, el norte geográfico está definido por el meridiano local en el que se encuentra. El meridiano local es el círculo máximo de la esfera celeste que pasa por el cenit del observador y por los polos celestes.

El sistema horizontal también se conoce como ``sistema altazimutal``. El módulo ``astropy.coordinates`` ofrece un marco potente para trabajar con objetos celestes.
